```python
import ipywidgets as widgets

upload_button = widgets.FileUpload()
display(upload_button)
```

```python
import cv2
import numpy as np

def read_uploaded_image(uploaded_image):
    img = cv2.imdecode(np.frombuffer(uploaded_image[-1]['content'], np.uint8), -1)
    return img
```

```python
import matplotlib.pyplot as plt

# Assuming 'uploaded_img' contains the uploaded image array
uploaded_img = read_uploaded_image(upload_button.value)
plt.imshow(cv2.cvtColor(uploaded_img, cv2.COLOR_BGR2RGB))
plt.show()
```
